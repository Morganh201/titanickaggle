```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.metrics import mean_absolute_error, classification_report, mutual_info_score
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.svm import SVC
import re
```

```python
df = pd.read_csv("train.csv")
X = df.copy()
Id = X.pop("PassengerId")
_ = X.pop("Name")
X.info()
```

```python
# Assume that nan prices are equal to 0
prices = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
X[prices] = X[prices].fillna(0)
X['Age'] = X["Age"].fillna(X["Age"].mean())
X['VIP'] = X['VIP'].fillna(False)
```

```python
# Change False to 0, True to 1
regex = r'([\d]+)'
X = X.replace(False, 0)
X = X.replace(True, 1)
X['Port'] = X['Cabin'].str.slice(start=-1)
temp = X['Cabin'].str.extract(regex, expand=False)
X['Num'] = pd.to_numeric(temp)
X['Deck'] = X['Cabin'].str.slice(start=0, stop=1)
X = X.drop("Cabin", axis=1)
X
```

```python
def onehot(encoder, df, column_name):
    encoder_df = pd.DataFrame(encoder.fit_transform(df[[column_name]]).toarray())
    encoder_df = encoder_df.add_prefix(f'{column_name}_')
    df = df.join(encoder_df)
    df = df.drop(column_name, axis=1)
    return df

encoder = OneHotEncoder(handle_unknown='ignore')
le = LabelEncoder()
X = onehot(encoder, X, "HomePlanet")
X = onehot(encoder, X, "Destination")
X = onehot(encoder, X, "Port")
X['Deck'] = le.fit_transform(X['Deck'])
X
```

```python
X['CryoSleep'] = X['CryoSleep'].fillna(np.random.randint(0, 1))
X['Num'] = X['Num'].fillna(X['Num'].mean())
plt.hist(X['Num'], density=True)
plt.show()
```

```python
y = X.pop("Transported")
X_train, val_X, y_train, val_y = train_test_split(X, y, random_state=0)
```

```python
# Logistic Regression

lr = LogisticRegression(random_state=0)
lr.fit(X_train, y_train)
predictions = lr.predict(val_X)
print(classification_report(predictions, val_y))

fpr, tpr, threshold = metrics.roc_curve(val_y, predictions)
auc = metrics.roc_auc_score(val_y, predictions)
plt.plot(fpr, tpr)
plt.title("ROC curve")
plt.ylabel("True Positive Rate")
plt.xlabel("False Positive Rate")
plt.show()

confusion_matrix = metrics.confusion_matrix(val_y, predictions)
display = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=[False, True])
display.plot()
plt.show()
```
